```python
responses = {

    "greeting": {
        "inputs": ["hi", "hello", "hey", "hii",
            "helllo"],

        "responses": [
            "Hello 👋",
            "Hey there!",
            "Hi student 🚀"
        ]
    },

    "ml": {
        "inputs": [
            "ml",
            "machine learning",
            "what is ml",
            "what is machine learning"
        ],

        "responses": [
            "Machine Learning is AI that learns from data.",
            "ML helps systems recognize patterns.",
            "ML helps systems improve from experience instead of explicit programming 🤖"
        ]

    },

    "robotics": {
        "inputs": [
            "robotics",
            "robot",
            "what is robotics"
        ],

        "responses": [
            "Robotics combines AI with machines.",
            "Robots use sensors and intelligent control."
        ]
    },

     "ai fact": {

        "inputs": [
            "ai fact",
            "tell me an ai fact",
            "fact"
        ],

        "responses": [
            "The term AI was officially coined in 1956 🧠",

            "ChatGPT predicts text one token at a time ⚡",

            "Neural networks were inspired by the human brain 🤯"
        ]
    },

    "bye": {
        "inputs": ["bye", "goodbye"],

        "responses": [
            "Goodbye 👋",
            "See you later!"
        ]
    }
}


st.sidebar.title("About")

st.sidebar.write(
    "EduNova AI is a beginner-friendly student AI assistant chatbot 🤖"
)


st.title("🤖 EduNova AI")

if st.button("Clear Chat"):

    st.session_state.chat_history = []

user_input = st.text_input("Ask something:")



if user_input:

    user_input = user_input.lower()

    response = None


    # MEMORY FEATURE

    if "my name is" in user_input:

        name = user_input.replace(
            "my name is",
            ""
        ).strip()

        st.session_state.memory["name"] = name

        response = f"Nice to meet you, {name}! 🚀"


    elif "what is my name" in user_input:

        if "name" in st.session_state.memory:

            response = (
                f"Your name is "
                f"{st.session_state.memory['name']} 👀"
            )

        else:

            response = (
                "I don't know your name yet 😭"
            )


    # FUZZY MATCHING

    else:

        found = False

        for category in responses:

            matches = get_close_matches(

                user_input,

                responses[category]["inputs"],

                n=1,

                cutoff=0.6
            )

            if matches:

                response = random.choice(
                    responses[category]["responses"]
                )

                found = True

                break




        if not found:

            unknown_responses = [

                "Interesting... I'm still learning that 🧠",

                "I haven't studied that topic yet 👀",

                "Teach me that sometime 😭"
            ]

            response = random.choice(
                unknown_responses
            )

    st.session_state.chat_history.append(
        ("You", user_input)
    )

    st.session_state.chat_history.append(
        ("EduNova", response)
    )
```

```python
for sender, msg in st.session_state.chat_history:

    if sender == "You":

        st.markdown(
            f"🧑 **You:** {msg}"
        )

    else:

        with st.spinner("Thinking..."):

            st.markdown(
                f"🤖 **EduNova:** {msg}"
            )
```
